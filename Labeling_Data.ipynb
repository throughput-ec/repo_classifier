# Merging ReadMe Files and Labels

```python
import pandas as pd
import json
import requests
import base64
```

```python
data = pd.read_csv('data/repository_readme_files.csv', index_col = 0)
```

```python
data.head(2)
```

```python
labels = pd.read_excel('data/GeologyPaleontologyFinalizedRandomSample.xlsx', index_col = 0, usecols = ['ID', 'URL', 'Purpose'])
```

```python
labels.head(2)
```

```python
labeled_data = pd.merge(data, labels, how="inner", left_on= 'repo_url', right_on='URL')
```

```python
labeled_data = labeled_data[['dacat', 'dacat_name', 'cr_item', 
                             'cr_name', 'URL', 'readme_content', 'Purpose']]
```

```python
labeled_data.to_csv('data/labeled_data.csv')
```
